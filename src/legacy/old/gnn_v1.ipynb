# GNN Dataset Exploration

This notebook explores the GNN dataset creation process for training Graph Neural Networks to predict optimal task placement in heterogeneous edge/IoT clusters.

## Dataset Overview
- **1,000 unique configurations** generated systematically
- **Parameters varied**: Network topology (0.05-0.50), replica configs (10 combinations), pre-warm levels (10 levels)
- **Realistic simulation data** with natural contention and queueing effects
- **Goal**: Train GNNs to learn optimal task→platform placements

## What we'll explore:
1. Dataset structure and available configurations
2. Sample simulation results and metrics
3. Feature extraction for GNN training
4. Graph representation of system states

```python
# Setup and imports
from pathlib import Path
import json
import pandas as pd
from typing import List, Dict, Any
import matplotlib.pyplot as plt
import seaborn as sns

# Configure paths
BASE_DIR = Path("/root/projects/my-herosim")
DATASETS_DIR = BASE_DIR / "simulation_data/gnn_datasets"
RESULTS_DIR = BASE_DIR / "simulation_data/initial_results_simple"

print("🔧 GNN Dataset Explorer Setup")
print(f"📁 Base directory: {BASE_DIR}")
print(f"📊 Datasets directory: {DATASETS_DIR.exists()}")
print(f"📈 Results directory: {RESULTS_DIR.exists()}")
print("✅ Ready to explore GNN datasets!")
```

```python
# 1. Dataset Structure Overview
def explore_dataset_structure(max_items: int = 5) -> None:
    """Explore the structure of generated GNN datasets"""
    if not DATASETS_DIR.exists():
        print("❌ No datasets directory found.")
        return
    
    ds_dirs = sorted(DATASETS_DIR.glob("ds_*"))
    print(f"🎯 Found {len(ds_dirs)} GNN dataset configurations")
    print(f"📋 Each dataset represents a unique system configuration:")
    print(f"   • Network topology (connection probability)")
    print(f"   • Replica configuration (clients/servers)")
    print(f"   • Pre-warm queue levels")
    print()
    
    # Show sample configurations
    for i, ds in enumerate(ds_dirs[:max_items]):
        print(f"📁 Dataset {i+1}: {ds.name}")
        
        # Check configuration file
        config_file = ds / "space_with_network.json"
        if config_file.exists():
            try:
                config = json.loads(config_file.read_text())
                network_prob = config.get("network", {}).get("connectionProbability", "N/A")
                print(f"   🌐 Network connection probability: {network_prob}")
            except:
                print("   ⚠️  Could not read config")
        
        # Check results index
        idx_file = ds / "results_index.txt"
        if idx_file.exists():
            lines = [ln.strip() for ln in idx_file.read_text().splitlines() if ln.strip()]
            print(f"   📊 Simulation results: {len(lines)} files")
            if lines:
                print(f"   📄 Sample result: {lines[0][:50]}...")
        else:
            print("   📊 No results yet")
        print()

explore_dataset_structure(100)
```
